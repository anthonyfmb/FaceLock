```python
import os
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
import cv2
```

```python
valid_user = "valid_user"
imposter = "imposter"
database = "database"
```

```python
count = 0
faces = []
y = []

print("imposter")
for imposter_img_name in os.listdir(database):
  print(imposter_img_name)
  if count < 500:
    imposter_img_dir = os.path.join(database, imposter_img_name)
    imposter_img = Image.open(imposter_img_dir).convert('L')
    imposter_img = imposter_img.resize((256, 256))

    img_numpy = np.array(imposter_img,'uint8')
    faces.append(img_numpy)
    y.append(0)
    count += 1
  else:
    break

print("valid")
count = 0
for valid_user_name in os.listdir(valid_user):
  print(valid_user_name)
  if count < 500:
    valid_user_dir = os.path.join(valid_user, valid_user_name)
    valid_img = Image.open(valid_user_dir).convert('L')
    valid_img = valid_img.resize((256, 256))

    img_numpy = np.array(valid_img,'uint8')
    faces.append(img_numpy)
    y.append(1)
    count += 1
  else:
    break

X_train, X_test, y_train, y_test = train_test_split(faces, y, test_size=0.1, random_state=42)
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.train(faces, np.array(y))
recognizer.write('classifier.yml') 


# correct = 0
# for i in range(len(X_test)):
#   y_pred, confidence = recognizer.predict(X_test[i])
#   if (y_pred != y_test[i]): print("diff")
#   else:
#     print("same")
#     correct += 1

# print(correct / len(X_test))

```
